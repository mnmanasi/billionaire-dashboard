```python
def demoComparison(df):
    gender_country_sum = df.groupby(['country', 'gender']).agg('sum')['finalWorth'].reset_index()
    gend_coun_sum_pivot = pd.pivot(gender_country_sum, index='country', columns='gender', values='finalWorth')
    return gend_coun_sum_pivot
```

```python
country_vs_gender = demoComparison(df)
country_vs_gender
```

```python
def demoComparison2(df):
    demo = df.groupby(['country', 'gender']).agg('sum')['finalWorth'].reset_index()
    demo_pivot = pd.pivot(demo, index='country', columns='gender', values='finalWorth')
    demo_pivot[demo_pivot['F'].isna()] = 0
    demo_pivot[demo_pivot['M'].isna()] = 0
    demo_pivot['total'] = demo_pivot['M'] + demo_pivot['F']
    top_demo_pivot = demo_pivot.nlargest(25, 'total')
    return demo, demo_pivot
```

```python
d, dp = demoComparison2(df)
dp
```

```python
import pandas as pd
import pycountry
import pycountry_convert as pc
```

```python
wdf = pd.read_csv('data/world-data-2023.csv')
wdf
```

```python
def get_continent(country_name):
    try:
        country_alpha2 = pycountry.countries.lookup(country_name).alpha_2   # get alpha-2 code
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)   # get continent code
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)   # map continent code to continent name
        return continent_name
    except (LookupError, KeyError):
        return 'Unknown'
```

```python
wdf['Continent'] = wdf['Country'].apply(get_continent)
wdf = wdf.dropna(subset=['GDP', 'Life expectancy'])
wdf
```
